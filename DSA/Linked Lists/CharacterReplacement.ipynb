Given a Linked List of characters. Write a python program to return a new string that is created by appending all the characters given in the linked list as per the rules given below.

**Rules** ->
1. Replace '*' or '/' by a single space
2. In case of two occurances of '*' or '/', replace those two occurances by a single space and convert the next character to upper case.

**Assumptions** ->
1. There will not be more than two consecutive occurances of '*' or '/'.
2. The linked list will always end with an alphabet.

**Sample Input** ->
A,n,* /,a,p,p,l,e, *,a,/,d,a,y, *, *,k,e,e,p,s,/, *,a,/,/,d,o,c,t,o,r, *,A,w,a,y

**Expected Output** ->
A,n, ,A,p,p,l,e, ,a, ,d,a,y, ,K,e,e,p,s, ,A, ,D,o,c,t,o,r, ,A,w,a,y

**Methodology**:

1. Single Character:
  1. if there is a single character: temp.data == '*'
  then, do temp.data = 'space'

2. Two characters in a row
  1. On encountering two characters in a row, replace the first character with a space and
  2. remove the other one from the list and
  3. make the next alphabet upper case

```python
class Node:

  def __init__(self, value):
    self.data = value
    self.next = None
```

```python
class LinkedList:

  def __init__(self):

    # Empty linked list: zero nodes: head = None
    self.head = None
    self.count = 0

  def __len__(self):
    return self.count

  # Inserting a value at the start of the linked list
  def insert_head(self, value):

    # adding a new node
    new_node = Node(value)

    # connecting he new node with the linked list
    new_node.next = self.head

    # reassigning the position of head to the new node
    self.head = new_node

    # incrementing the length of nodes
    self.count = self.count + 1

  # Printing the linked list
  def __str__(self):

    curr = self.head

    result = ''

    while curr != None:

      result = result + str(curr.data)
      curr = curr.next

    return result[:-3]

  # Insertion at tail: appending the linked list
  def append(self, value):

    new_node = Node(value)

    # if the linked list is EMPTY
    if self.head == None:
      self.head = new_node
      self.count = self.count + 1
      return

    curr = self.head

    while curr.next != None:
      curr = curr.next

    # at the last node
    curr.next = new_node
    self.count = self.count + 1

  # Inserting after a certain value
  def insert_after(self, after, value):

    new_node = Node(value)

    curr = self.head

    while curr != None:

      if curr.data == after:
        break

      curr = curr.next

      # case 1: Break: We found the after node: curr -> not None
    if curr != None:
      new_node.next = curr.next
      curr.next = new_node
      curr = new_node
      self.count = self.count + 1
    else:
      return "After node not found!!!"
      # case 2: No break: Loop ran till end, after node not found: curr -> Nonee

  # Deleting whole linked list values
  def clear(self):
    self.head = None
    self.count = 0

  # Deleting at head
  def delete_head(self):

    if self.head == None:
      # Empty List
      return 'Empty Linked List'
    self.head = self.head.next
    self.count = self.count - 1

  # Deleting at tail
  def pop(self):

    if self.head == None:
      # Empty List
      return 'Empty Linked List'

    curr = self.head

    # if there is only one node in linked list
    if curr.next == None:
      #it is head
      return self.delete_head()

    while curr.next.next != None:
      curr = curr.next

    # We are at 2nd last node
    curr.next = None
    self.count = self.count - 1

  # Removing a certain node
  def remove(self, value):

    if self.head == None:
      # Empty List
      return 'Empty Linked List'

    if self.head.data == value:
      # deleting Head node
      return self.delete_head()

    curr = self.head

    while curr.next != None:
      if curr.next.data == value:
        break
      curr = curr.next

    # Value not found
    if curr.next == None:
      return 'Not Found'
    else:
      # Value Found
      curr.next = curr.next.next
      self.count = self.count - 1

  # Searhing by Value
  def search(self, value):

    curr = self.head
    pos = 0

    while curr != None:
      if curr.data == value:
        return 'Found at index: ', pos
      curr = curr.next
      pos = pos + 1

    return 'Not Found'

  # Search by index
  def __getitem__(self, index):

    curr = self.head
    pos = 0

    while curr != None:
      if pos == index:
        return curr.data
      curr = curr.next
      pos = pos + 1

    return 'Not Found'

  # Replacing the maximum value of LL
  def replace_max(self, value):

    temp = self.head
    max = temp

    while temp != None:
      if temp.data > max.data:
        max = temp
      temp = temp.next

    max.data = value

  # Sum of integers at Odd index
  def OddPosSum(self):

    temp = self.head
    pos = 0
    result = 0

    while temp != None:
      if pos % 2 != 0:
        print("Index: ",pos, "Data: ",temp.data)
        result = result + temp.data

      pos += 1
      temp = temp.next

    return "sum = ", result

  # Inplace reversal of LL
  def Reverse(self):

    prev_node = None # Previos node
    curr_node = self.head

    while curr_node != None:
      next_node = curr_node.next # saving the address of next node
      curr_node.next = prev_node
      # reversing the link from prev_node -> curr_node
      # to prev_node <- curr_node, next_node stores address of successive node.
      prev_node = curr_node # moving forward to reverse another node
      curr_node = next_node

    self.head = prev_node

  #changing characters
  def char_change(self):

    temp = self.head

    while temp != None:

      # for single character
      if temp.data == '*' or temp.data == '/':
        temp.data = ' '

        # for two consecutive characters
        if temp.next.data == '*' or temp.next.data == '/':
          temp.next.next.data = temp.next.next.data.upper()
          temp.next = temp.next.next

      temp = temp.next


```
